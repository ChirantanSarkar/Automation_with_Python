```python
from io import StringIO
from fake_useragent import UserAgent
import pandas as pd
import requests

url="https://en.wikipedia.org/wiki/List_of_Star_Wars_films"
headers={"User-Agent": UserAgent().random} #Mimics as a human using a browser
download=requests.get(url,headers=headers) #Downloads the data from web
data=StringIO(download.text) #Turns it into readable text
df=pd.read_html(data)
len(df)
```
